## Trabajo Final Integrador - Análisis del Comportamiento de Usuarios de Streaming

Este notebook consolida los hallazgos del **Análisis Exploratorio de Datos (EDA)** y del **Análisis de Componentes Principales (PCA)** realizados sobre el dataset de 8,000 usuarios de streaming en Latinoamérica. A partir de la evidencia estadística directa de los análisis previos, se estructuran las conclusiones del comportamiento del usuario, el diagnóstico operativo y las recomendaciones estratégicas para la toma de decisiones empresariales.

---
### Objetivos Analizados:
1. **Comprender los perfiles de consumo e interacción** demográfica de los usuarios.
2. **Evaluar el rendimiento del soporte técnico** y determinar cuellos de botella geográficos u operativos.
3. **Analizar la efectividad de la estructura de planes** comerciales en términos de engagement.
4. **Validar la viabilidad de una reducción dimensional (PCA)** y evaluar la agrupación natural de los usuarios según su plan de suscripción.

## 1. Hallazgos Clave del Análisis Exploratorio de Datos (EDA)

El EDA nos permitió caracterizar de forma univariada, bivariada y multivariada a la base de usuarios de la plataforma, arrojando luz sobre las siguientes realidades operativas y de consumo:

### A. Perfil Demográfico y Hábitos de Consumo
* **Demografía Productiva:** El usuario promedio tiene **35.6 años** (mediana 35), con un núcleo principal (IQR) concentrado entre los **29 y 41 años**. Este es un público maduro, económicamente independiente y con capacidad de consumo sostenido.
* **Intensidad de Consumo:** El tiempo de visualización mensual promedia **768.4 minutos** (~12.8 horas/mes, es decir, unos 25 minutos diarios), pero con una dispersión amplia (Std = 507.9 mins) y la existencia de un segmento de superusuarios que superan los **4,000 minutos mensuales** (~66 horas/mes).
* **Diversidad de Preferencias:** Excluyendo un mínimo 2.96% de registros "Desconocidos", los géneros favoritos están perfectamente balanceados (~13.8% cada uno para Comedia, Drama, Romance, Documental, Thriller, Acción y Crimen), lo que obliga a mantener un catálogo horizontal y diverso.

### B. Evaluación de la Operativa y Soporte
* **Carga Operativa Controlada:** El promedio de tickets de soporte técnico por usuario es muy bajo (**0.79 tickets**), y el 75% de los usuarios tiene como máximo 1 ticket mensual. Esto indica una plataforma sumamente estable en términos técnicos y de facturación.
* **Estabilidad Continental:** El soporte promedio es geográficamente uniforme, oscilando entre **0.77 (Brasil/Colombia)** y **0.82 (Argentina)**. No existen inestabilidades de red o distribución de contenido (CDN) específicas de un país que disparen las quejas en mercados particulares.

### C. Eficacia del Modelo de Planes
* **Validación de la Oferta Comercial:** El plan de suscripción contratado demostró ser la variable más determinante del volumen de consumo. Existe un incremento del **72.4% en el tiempo de visualización** al pasar del plan Básico (promedio **591.6 mins**) al Premium (promedio **1,019.8 mins**). Los planes superiores logran su objetivo de capturar un mayor engagement, lo que a su vez se traduce en un menor riesgo de cancelación (*churn*).
* **Independencia Demográfica:** El análisis multivariado evidenció que la edad no se correlaciona con el tiempo de consumo ($r = 0.01$) ni con las quejas ($r = 0.002$). Un cliente de 60 años en el plan Premium consume tanto contenido y genera tan pocos tickets como uno de 20 años en el mismo plan. La propuesta de valor de los planes es universalmente efectiva.

## 2. Hallazgos Clave del Análisis de Componentes Principales (PCA)

El PCA aportó una perspectiva matemática y geométrica sobre las relaciones subyacentes de las tres variables numéricas estandarizadas (`age`, `monthly_watch_time_mins`, y `customer_support_tickets`):

### A. Diagnóstico de la Reducción Dimensional
* **Varianza Explicada Plana:** El Scree Plot muestra que los tres componentes explican casi la misma varianza: PC1 (**33.74%**), PC2 (**33.41%**), y PC3 (**32.85%**).
* **Inefectividad de la Reducción:** Debido a que la correlación lineal entre las tres variables es prácticamente nula ($r pprox 0$), el espacio de datos es esférico. Reducir la dimensionalidad de 3 variables a 2 componentes principales (PC1 y PC2) conserva el **67.15%** de la varianza, perdiendo un **32.85%** de la información. Esta pérdida es casi idéntica a descartar directamente una variable física completa (33.3%). Por ende, **el PCA no es un método efectivo de compresión de información para este dataset**.

### B. Estructura de los Componentes
* **PC1 ("Alto Valor, Menor Fricción"):** Dominado positivamente por `monthly_watch_time_mins` (0.735) y `age` (0.564), y negativamente por `customer_support_tickets` (-0.375). Representa a los usuarios adultos estables que consumen de forma activa y casi no generan incidencias técnicas.
* **PC2 ("Fricción en Edad Avanzada"):** Dominado fuertemente por `customer_support_tickets` (0.800) y `age` (0.597), con peso nulo en consumo. Identifica a usuarios mayores que experimentan problemas técnicos o de facturación frecuentes, independientemente de su nivel de uso.

### C. Superposición de los Planes de Suscripción
* **Solapamiento Masivo:** Al proyectar PC1 vs. PC2, las coordenadas de los usuarios de los planes Básico, Estándar y Premium se solapan de manera casi total en una gran nube homogénea. Aunque el promedio de PC1 se desplaza de forma sutil según el plan (Básico: -0.261 vs. Premium: 0.355) debido a los diferentes consumos de minutos, la distribución de los tres planes cubre rangos idénticos.
* **Implicación Analítica:** **El plan contratado no define un perfil demográfico o conductual exclusivo**. Los hábitos individuales de consumo y reclamo varían tanto dentro de cada plan que imposibilitan separar a los usuarios en categorías limpias usando únicamente estos criterios. La segmentación basada en planes comerciales no refleja la estructura conductual real de los usuarios.

## 3. Recomendaciones Estratégicas y de Negocio

Basados en los resultados numéricos y analíticos anteriores, se plantean las siguientes acciones corporativas:

### 1. Campañas de Upselling Dirigidas (Foco en el Plan Básico)
* **Contexto:** El **45%** de la base de usuarios pertenece al plan Básico (con un consumo medio inferior de ~592 mins/mes).
* **Acción:** Diseñar incentivos de migración hacia los planes Estándar y Premium dirigidos a usuarios del plan Básico cuyo consumo actual supere la media de su grupo. Un usuario Básico con alto consumo es el candidato ideal para ser migrado a un plan superior (Estándar/Premium), aumentando su ARPU y asegurando su retención al dotarlo de mejor calidad de video y pantallas adicionales.

### 2. Optimización Localizada de Soporte Premium (Chile y Perú)
* **Contexto:** En Chile y Perú, los usuarios Premium presentan un promedio de tickets de soporte superior a los del plan Básico (Chile: **0.92 vs. 0.76**; Perú: **0.89 vs. 0.76**).
* **Acción:** Auditoría de las llamadas y tickets en estos países para identificar problemas en características exclusivas de las cuentas Premium (como transmisión UHD/4K, configuraciones de Smart TVs o cobros multihogar). Se debe entrenar específicamente a los equipos de soporte de Chile y Perú en estas funcionalidades avanzadas para reducir los tiempos de resolución y evitar la fuga de clientes Premium de alto valor.

### 3. Transición a una Segmentación de Clientes Avanzada
* **Contexto:** La segmentación comercial actual (Básico, Estándar, Premium) presenta un solapamiento completo en el espacio tridimensional de edad, visualización y soporte (PCA), mostrando que el plan de suscripción no es un descriptor preciso del comportamiento de uso real.
* **Acción:** Adoptar una metodología de segmentación conductual basada en **clustering no supervisado (como K-Means)** en lugar de segmentación comercial. Este algoritmo permitirá agrupar a los usuarios en clústeres reales basados en afinidad por géneros, distribución horaria de visualizaciones e inicios de sesión, permitiendo una personalización del catálogo y campañas de marketing de fidelización mucho más efectivas.

## 5. Limitaciones y Mejoras Futuras (Calidad de Datos)

### A. Limitación de Origen: Detección de Outliers Temporales
Durante la fase de auditoría e inspección del dataset, se identificó una anomalía en el campo temporal de última conexión (`last_login_date`):
* **Hallazgo:** Un total de **15 registros** (equivalentes al **0.18%** de la muestra total de 8,000 usuarios) registran como fecha de último inicio de sesión el **`2029-01-01`**.
* **Diagnóstico:** Dado que esta fecha es idéntica en todos los casos y se sitúa en el futuro respecto a la fecha actual de análisis (28/06/2026), se concluye empíricamente que responde a un **error de registro del sistema**. Específicamente, apunta a un valor por defecto (*default value*) inyectado de forma automatizada por el backend de la base de datos o el proceso de extracción, probablemente ante la presencia de datos originales faltantes, corruptos o con problemas de codificación.
* **Impacto en el Estudio:** Debido a su nula representatividad estadística (menos del 0.2% de la muestra), esta inconsistencia **no altera las conclusiones generales** derivadas del EDA ni altera las proyecciones geométricas del PCA. Sin embargo, se documenta formalmente como una limitación menor de calidad de datos en el origen del dataset.

### B. Mejora Futura: Reglas de Validación en el Pipeline de Limpieza
Para robustecer la calidad de los datos en iteraciones futuras del proyecto, se propone:
* **Validación Temporal Dinámica:** Incorporar en el script de curación de datos ([02_calidad_y_limpieza.ipynb](file:///home/bucci/Proyectos/PI_Mineria_Datos_1/notebooks/02_calidad_y_limpieza.ipynb)) una regla de validación temporal condicional que compare cada fecha con el tiempo de ejecución actual (`current_date = datetime.now()`).
* **Tratamiento Automatizado:** Ante la detección de registros donde `last_login_date > current_date`, el pipeline de limpieza deberá:
  1. Aislar los registros en un archivo de logs de auditoría para reportar al equipo de ingeniería de datos.
  2. Imputar dichos valores anómalos de forma controlada (por ejemplo, reemplazándolos con el valor nulo de tipo `NaT` de Pandas o la última fecha de conexión válida del usuario) para que no interfieran en las métricas de uso y modelos analíticos.
